### Задача 2
В ящике 5 апельсинов и 4 яблока. Наудачу выбираются 3 фрукта. Какова вероятность, что все три фрукта – апельсины?

В интернете полученный аналитически ответ 0.119. Подтверждается ли он эксперементально?


```python
# Ваше решение
```

```python
import random
```

```python
#Создаем ящик для фруктов
```

```python
Box = ["orange"]*5 + ["apple"]*4
```

```python
#задаем число экспириментов
```

```python
n = 1000000
```

```python
# формируем выборку
```

```python
A = pd.DataFrame({'A1':list(map(lambda a: random.choice(Box), range(n))),'A2':list(map(lambda a: random.choice(Box), range(n))), 'A3':list(map(lambda a: random.choice(Box), range(n)))})
```

```python
A
```

```python
# считаем вероятность того, что будут выбраны 3 апельсина 
```

```python
Probability = len(A.query('A1 == "orange" & A2 == "orange" & A3 == "orange"'))/A.shape[0]    
```

```python
Probability
```

```python
#Полученный результат немного отличается от аналитеческого ответа, но погрешность не значительная
```

### Задача 3
Мастер, имея 10 деталей, из которых 3 – нестандартных, проверяет детали одну за другой, пока ему не попадется стандартная. Какова вероятность, что он проверит ровно две детали?


В интернете полученный аналитически ответ 7/30 или 0.23333. Подтверждается ли он эксперементально?

```python
# Ваше решение
```

```python
#Создаем ящик для деталей
```

```python
Box = ["standart"]*7 + ["nonstandart"]*3
```

```python
#задаем число экспириментов
```

```python
n = 1000000
```

```python
# формируем выборку
```

```python
A = pd.DataFrame({'A1':list(map(lambda a: random.choice(Box), range(n))),'A2':list(map(lambda a: random.choice(Box), range(n)))})
```

```python
A
```

```python
# считаем вероятность того, что мастер проверит 2 детали
```

```python
Probability = len(A.query('A1 == "nonstandart" & A2 == "standart"'))/A.shape[0] 
```

```python
Probability
```

```python
#Полученный результат немного отличается от аналитеческого ответа, но погрешность не значительная
```
